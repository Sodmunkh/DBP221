```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data0 = pd.read_excel(r'D:\4th Semester\DBP221 - Python\COURSEWORK\Бие даалт - 1.xlsx', sheet_name = 'Ирц')
data1 = pd.read_excel(r'D:\4th Semester\DBP221 - Python\COURSEWORK\Бие даалт - 1.xlsx', sheet_name = 'Семинар')
data2 = pd.read_excel(r'D:\4th Semester\DBP221 - Python\COURSEWORK\Бие даалт - 1.xlsx', sheet_name = 'Шалгалт')
data3 = pd.read_excel(r'D:\4th Semester\DBP221 - Python\COURSEWORK\Бие даалт - 1.xlsx', sheet_name = 'Бие даалт')
data4 = pd.read_excel(r'D:\4th Semester\DBP221 - Python\COURSEWORK\Бие даалт - 1.xlsx', sheet_name = 'Хүмүүжил хандлага')
data5 = pd.read_excel(r'D:\4th Semester\DBP221 - Python\COURSEWORK\Бие даалт - 1.xlsx', sheet_name = 'Нэмэлт оноо')
data6 = pd.read_excel(r'D:\4th Semester\DBP221 - Python\COURSEWORK\Бие даалт - 1.xlsx', sheet_name = 'Нийт')
```

```python
#Ирц
data0_cl = data0.replace(['и','т'],[1,0])
data0_sum = data0_cl.iloc[:, 4:].sum(axis = 1)
data0_pt = ((data0_sum / 32)*10).round()
```

```python
#Семинар
data1_cl = data1.fillna(0)
data1_sum = data1_cl.iloc[:, 4:].sum(axis = 1)
data1_pt = (data1_sum * (25/19)).round()
```

```python
#Шалгалт
data2_cl = data2.fillna(0)
data2_sum = (data2_cl.iloc[:, 4:].sum(axis = 1)).round()
```

```python
#Бие Даалт
data3_cl = data3.fillna(0)
data3_sum = (data3_cl.iloc[:, 4:].sum(axis = 1)).round()
```

```python
#Хүмүүжил хандлага
data4_cl = data4.fillna(0)
data4_sum = (data4_cl.iloc[:, 4:].sum(axis = 1)).round()
```

```python
#Нэмэлт оноо
data5_cl = data5.fillna(0)
data5_sum = (data5_cl.iloc[:, 4:].sum(axis = 1)).round()
```

```python
#Оюутны код болон нэрс
student = data6.iloc[:-1, 2:4]
```

```python
#Бүх нийлбэр
niilber = (data0_pt + data1_pt + data2_sum + data3_sum + data4_sum + data5_sum).round()
```

```python
niit = pd.DataFrame({
    'Ирц' : data0_pt,
    'Семинар' : data1_pt,
    'Шалгалт' : data2_sum,
    'Бие Даалт' : data3_sum,
    'Хүмүүжил хандлага' : data4_sum,
    'Нэмэлт оноо' : data5_sum,
    'Нийт' : niilber
})
```

```python
#Бүхэл dataframe болгосон
niit1 = pd.concat([student, niit], axis = 1)
```

```python
conditions = [
    (niit1['Нийт'] >= 96) & (niit1['Нийт'] <=100),
    (niit1['Нийт'] >= 90) & (niit1['Нийт'] <=95),
    (niit1['Нийт'] >= 87) & (niit1['Нийт'] <=89),
    (niit1['Нийт'] >= 83) & (niit1['Нийт'] <=86),
    (niit1['Нийт'] >= 80) & (niit1['Нийт'] <=82),
    (niit1['Нийт'] >= 77) & (niit1['Нийт'] <=79),
    (niit1['Нийт'] >= 73) & (niit1['Нийт'] <=76),
    (niit1['Нийт'] >= 70) & (niit1['Нийт'] <=72),
    (niit1['Нийт'] >= 67) & (niit1['Нийт'] <=69),
    (niit1['Нийт'] >= 63) & (niit1['Нийт'] <=66),
    (niit1['Нийт'] >= 60) & (niit1['Нийт'] <=62),
    (niit1['Нийт'] >=  0) & (niit1['Нийт'] <=59)
]
```

```python
values = ['A', 'A-',
          'B+', 'B', 'B-',
          'C+', 'C', 'C-',
          'D+', 'D', 'D-',
          'F']
```

```python
gpa = ['4.0', '3.7',
          '3.4', '3.0', '2.7',
          '2.4', '2.0', '1.7',
          '1.3', '1.0', '0.7',
          'Унасан']
```

```python
niit1['Үсгэн үнэлгээ'] = np.select(conditions, values)
niit1['Голч дүн'] = np.select(conditions, gpa)
```
